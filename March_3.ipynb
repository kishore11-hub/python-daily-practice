```python
#Number Guessing Game

import random


secret_number = random.randint(1, 100)


print("Guess the Number Between 1 to 100")
attempt = 0
max_attempt = 7

while attempt < max_attempt:
    user = int(input("Guess My Number :"))
    if user <= 0 or user > 100:
        print("Guess the Number Between 1 - 100")
    attempt = attempt + 1

    if user == secret_number:
        print("correct, You Guessed My Number in", attempt ,"Attempts")
        break

    if user >= secret_number:
        print("My Number is low")

    else:
        print("My Number is High")

else:
    print("You have Attempted Maximum Limits")
    print("My Number is", secret_number)
    
```
