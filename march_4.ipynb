```python
#password strength checker

password = input("Enter the Password:") 

has_upper = False
has_lower = False
has_number = False
has_special = False

for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_number = True
    else:
        has_special = True

if len(password) >= 8 and has_upper and has_lower and has_special and has_number:
    print("strong Password")

else:
    print("Weak Password")
```
