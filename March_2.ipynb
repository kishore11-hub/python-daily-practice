```python
# 1. Reverse a Number 

num = int(input("enter the number:"))
result = 0

while num > 0:
    d = num % 10
    result = result * 10 + d
    num = num // 10

print (result)
    
```

```python
# 4. check for number palindrome or not

num = int(input("Enter the number:"))
num1 = num
result = 0

while num > 0:
    d = num % 10
    result = result * 10 + d
    num = num // 10
if result == num1:
    print("palindrome")

else:
    print("Not a palindrome")

# string palindrome or Not

text = input("enter:")
result = ""

for i in text:
    result = i + result

if result == text:
    print("palindrome")
else:
    print("not a palindrome")
```

```python
# 5. Factorial of a Number

num = int(input("Enter the number:"))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print(fact)

# while Loop

num = int(input("Enter the number:"))
fact = 1
i = 1

while i <= num:
    fact = fact * i
    i = i + 1
    

print(fact)
```

```python
# 6. prime number 

num = int(input("Enter the Number:"))

if num <= 1:
    print("not a prime")

elif num == 2:
    print("Prime")

else:
    for i in range(2, num):
        if num % i == 0:
            print("not a prime")
            break
    else:
        print("Prime")
 
```
